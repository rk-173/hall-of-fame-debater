This is to examine some of the functionality of the NBA API, and to identify some functions we'll be using as tools for the agents

```python
import nba_api
```

```python
from nba_api.stats.static import players
player_name = "Dirk Nowitzki"

player_dict = players.get_players()

player = [p for p in player_dict if p["full_name"] == player_name][0]
player_id = player["id"]
player_id
```

```python
from nba_api.stats.endpoints import playerawards

awards = playerawards.PlayerAwards(player_id=player_id)
df = awards.get_data_frames()[0]
df
```

```python
# Aggregate the Awards for the player
summary = df['DESCRIPTION'].value_counts().to_dict()

# Sporting news MVP and olympic appearances don't matter for this
# Since this is NBA focused and sporting news MVP stops at 2011
summary.pop('NBA Sporting News Most Valuable Player of the Year', None)
summary.pop('Olympic Appearance', None)

summary
```
